```python
from google.colab.patches import cv2_imshow
import cv2
image = cv2.imread('/content/caat.png')
cv2_imshow(image)
```

```python
print("shape:", image.shape) # height , width, color channels (Red Green Blue)

print("H:", image.shape[0])
print("W:", image.shape[1])
print("C:", image.shape[2])
```

```python
(b, g, r)= image[100, 50] # bu 100 - qatordagi 50 - ustundagi ranglarni olib beradi
print("Blue", b, "Green", g, "Red", r)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
resize = cv2.resize(image, (200, 200))
cv2_imshow(resize)
# rasm olchamini o'zgartira olamiz
```

```python
cropped = image[100:300, 150:400]
cv2_imshow(cropped)
```

```python
roteted = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(roteted)
```

```python
flipped = cv2.flip(image, 1)
cv2_imshow(flipped)
```

```python
cv2.imwrite("newimage.jpg", flipped)
```

```python
cropped_cat = cv2.imread("/content/newimage.jpg")
cv2_imshow(cropped_cat)
```
